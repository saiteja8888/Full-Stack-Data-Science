```python
def check_score(lst):
    d={'#':5,'O':3,'X':1,'!':-1,'!!':-3,'!!!':-5}
    s=0
    for i in lst:
        if type(i)==list:
            for j in i:
                if j in d:
                    s+=d[j]
        elif i in d:
            s+=i.d[j]
    return s

print(check_score([["#","!"],["!!","X"]]))
print(check_score([["!!!","O","!"],["X","#","!!!"],["!!","X","O"]]))
```

```python
def combinations(*args):
    from functools import reduce
    com=reduce (lambda a,b:a*b,list(args))
    return com
     
print(combinations(2,3))
print(combinations(3,7,4))
print(combinations(2,3,4,5))
```

```python
def prime(num):
    if num==0 or num==1:
        return False
    else:
        for i in range(2,int(num**0.5)+1):
            if num%i==0:
                return False
    return True

print(prime(7))
print(prime(56963))
print(prime(5151512515524))
```

```python
def to_boolean_list(text):
    text.lower()
    lst=[]
    for i in text:
        temp=ord(i)-64
        temp%=2
        lst.append(True) if temp==1 else lst.append(False)
    return lst

print(to_boolean_list("deep"))
print(to_boolean_list("loves"))
print(to_boolean_list("tesh")) 
```
